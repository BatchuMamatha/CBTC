```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Reading the Dataset
df = pd.read_excel(r"C:\Users\Poojitha\Downloads\Unemployment_Rate_upto_11_2020.xlsx")
```

```python
# Data Inspection
print(df.head())
print(df.tail())
print(df.shape)  # Get the number of rows and columns
print(df.info())  # Get a concise summary of the DataFrame
print(df.isnull().sum())  # Check for null values
print(df.describe())  # Get descriptive statistics
```

```python
# Renaming Columns
df.columns = ["States", "Date", "Frequency", "Estimated Unemployment Rate", 
              "Estimated Employed", "Estimated Labour Participation Rate", 
              "Region", "longitude", "latitude"]
```

```python
# Data Visualization

# Unemployment Rate Analysis by Region
plt.title("Indian Unemployment - Estimated Employed by Region")
sns.histplot(x="Estimated Employed", hue="Region", data=df)
plt.show()
```

```python
# Unemployment Rate by Region
plt.figure(figsize=(12, 10))
plt.title("Indian Unemployment - Unemployment Rate by Region")
sns.histplot(x="Estimated Unemployment Rate", hue="Region", data=df)
plt.show()
```

```python
# Exploratory Data Analysis

# Pie Chart of Employment Statistics
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
labels = ['Estimated Unemployment Rate', 'Estimated Employed', 'Estimated Labour Participation Rate']
sizes = [50, 50, 50]
ax.pie(sizes, labels=labels, autopct='%1.2f%%')
plt.show()
```

```python
# Histograms of All Features
sns.set(style="darkgrid")
df.hist(bins=20, figsize=(20, 15))
plt.show()
```

```python
# Count Plot of Regions
sns.countplot(x='Region', data=df)
plt.title("Count of Records by Region")
plt.show()
```

```python
# Scatter Plots
colors = ['Red', 'Blue', 'Green', 'Orange', 'Purple']
regions = ['South', 'East', 'West', 'North', 'Northeast']
```

```python
# Estimated Unemployment Rate vs Estimated Employed
plt.figure(figsize=(10, 6))
for i in range(len(regions)):
    region_df = df[df['Region'] == regions[i]]
    plt.scatter(region_df['Estimated Unemployment Rate'], region_df['Estimated Employed'], 
                c=colors[i], label=regions[i])
plt.xlabel("Unemployment Rate")
plt.ylabel("Estimated Employed")
plt.legend()
plt.title("Unemployment Rate vs Estimated Employed")
plt.show()
```

```python
# Estimated Employed vs Estimated Labour Participation Rate
plt.figure(figsize=(10, 6))
for i in range(len(regions)):
    region_df = df[df['Region'] == regions[i]]
    plt.scatter(region_df['Estimated Employed'], region_df['Estimated Labour Participation Rate'], 
                c=colors[i], label=regions[i])
plt.xlabel("Estimated Employed")
plt.ylabel("Estimated Labour Participation Rate")
plt.legend()
plt.title("Estimated Employed vs Estimated Labour Participation Rate")
plt.show()
```

```python
# Correlation Matrix
numeric_df = df.select_dtypes(include=[np.number])  # Select only numeric columns
corr_matrix = numeric_df.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
